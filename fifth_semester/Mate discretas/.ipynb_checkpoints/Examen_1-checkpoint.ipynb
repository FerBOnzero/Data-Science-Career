##  $\color{blue}{\text {Inciso a) Términos divisibles por 7}}$


### $\color{violet}{\text {Solución: }}$

Para la solución de este ejercicio, primero simulemos utilizando python algunos valores de la sucesiónde fibonacci para saber si encontramos algún patrón que nos diga cuáles términos son divisibles por 7


```python
import numpy as np
```

Definimos nuestra función que nos regrese una lista con los n términos de Fibonacci

```python
'''
Función que regresa la lista con los n términos de la sucesión de Fibonacci.

Parámetros:
n := cantidad de valores a calcular por la sucesión de Fibonacci
'''

def fibo(n):
    fibo_list = [0,1]
    for i in range(n-1):
        fibo_list.append(fibo_list[i] + fibo_list[i+1])
    return fibo_list
```

Ahora simulemos sobre los primeros 100 términos de la sucesión de Fibonacci

```python
for i in range(100):
    if fibo(101)[i] % 7 == 0:
        print(f'at position {i} found term of fibo {fibo(100)[i]} divisible by 7')
```

Como pudimos ver anteriormente, vemos que para $n\in\mathbb{N}$ múltiplo de $8$ obtenemos que $F_{n}$ es divisible por $7$. Por lo tanto procederemos por inducción sobre este supuesto:
1. Caso base: $n = 1$\
    Supongamos $n=0=8\cdot m \text{ con }m\in\mathbb{N}$, entonces tenemos que el término 1 de la sucesión de Fibonacci es 0, por lo tanto queda claro que 0 es divisible por 7, de este modo se cumple para el caso base. 
2. Hipótesis de inducción:\
    Supongamos que se cumple para $n=k = 8\cdot m \text{ con }m\in\mathbb{N-0}$ que el término $F_n$ es divisible por $7$.
3. Paso inductivo:\
    Por demostrar que se cumple para $n = k + 8 = 8 \cdot (m+1) \text{ con }m\in\mathbb{N-0}$, ya que si decimos "Por demostrar que se cumple para $n=k+1 \Rightarrow k+1 = 8 \cdot m + 1$" en donde tendríamos una contradicción inmediata, ya que $8 \cdot m + 1$ ya no es múltiplo de $8$.\
    Por hipótesis inductiva tenemos tenemos que es válido para $k$, entonces podemos ver a $k$ como $k=8\cdot m, \text{ con }m\in\mathbb{N-0}$, es decir, tenemos que $F_{k} = F_{8\cdot m} | 7$ (es decir, el k-ésimo término es divible por 7). Ahora procedemos a desarrollar la expresión $$F_{k+8}=F_{8\cdot (m+1)} = F_{8\cdot m + 8} = F_{8\cdot m + 7} + F_{8\cdot m + 6} = (F_{8\cdot m + 6} + F_{8\cdot m + 5}) + F_{8\cdot m + 6}= 2 \cdot F_{8\cdot m+ 6} + F_{8\cdot m + 5 } = 2\cdot (F_{8\cdot m + 5}+ F_{8\cdot m + 4}) + F_{8\cdot m + 5} = 3\cdot F_{8\cdot m + 3} + 5\cdot F_{8\cdot m + 4} = 5\cdot F_{8\cdot m + 3} + 5\cdot F_{8\cdot m + 2} + 3\cdot F_{8\cdot m + 3} = 8\cdot F_{8\cdot m + 3}+ 5\cdot F_{8\cdot m +2} = 8\cdot F_{8\cdot m + 2} + 8\cdot F_{8\cdot m + 1} + 5\cdot F_{8\cdot m + 1} = 8\cdot F_{8\cdot m + 1} + 13\cdot F_{8\cdot m + 2} = 8\cdot F_{8\cdot m + 1} + 13\cdot F_{8\cdot m + 1} + 13\cdot F_{8\cdot m } = 21\cdot F_{8\cdot m + 1}+ 13\cdot F_{8\cdot m }$$ Por hipótesis de inducción, tenemos que $F_{8\cdot m }$ es divisible por 7, por lo tanto cualquier múltiplo de $F_{8\cdot m }$ también lo es, además veamos a $21\cdot F_{8\cdot m + 1}$ como $7(3\cdot F_{8\cdot m + 1})$, entonces este también es múltiplo de 7, por lo tanto para $n\in\mathbb{N}$ múltiplo de 8, $F_{n}$ es divisible por 7. $$\blacksquare$$

##  $\color{blue}{\text {Inciso b) Demostrar fórmula para sucesión Fibonacci}}$


### $\color{violet}{\text {Solución: }}$


Consideremos la siguiente ecuación cuadrática: $$x^2 - x - 1 = 0$$ la cual tiene raíces $$x_1 = \frac{1+\sqrt{5}}{2} \text{ y } x_2 = \frac{1-\sqrt{5}}{2}$$ Para poder demostrar el problema del inciso, ocuparé el siguiente resultado, el cuál es:

$\color{red}{\text{Para cualquier solución }x \text{ de } x^2−x−1=0, \text{ }x^n=xF_n+F_{n−1},\quad n\geq 1}$\
$\color{violet}{\text{Demostración lema}}$ 
La demostración se hará por inducción:\
$\color{violet}{\text{caso base n = 1}}$ \
Se cumple la igualdad, ya que $x^1 = x\cdot 1 + 0 = x$\
$\color{violet}{\text{hipótesis de inducción n = k}}$ \
Supongamos que se cumple para alguna k natural de modo que $x^k=xF_k+F_{k−1}$
$\color{violet}{\text{paso de inductivo, por demostrar que se cumple para n = k + 1}}$ \
Multipliquemos ambos lados de la igualdad de la hipótesis de inducción por $x$ y desarrollemos, es decir: 
$$x^{k+1} = x^2 F_k + xF_{k-1} = (x+1)F_k + xF_{k-1} = x(F_k + F_{k-1}) + F_k = xF_{k+1} + F_k$$ $$\blacksquare$$

Una vez demostrado esto, procedemos a utilizar el resultado con las soluciones $x_1$ y $x_2$, es decir: 
$$x_1^n = x_1 F_{n} + F_{n-1}$$ $$x_2^n = x_2 F_{n} + F_{n-1}$$ ahora restemos ambas expresiones: $$(x_1^n - x_2^n) = (x_1 - x_2)F_n \Rightarrow \frac{x_1^n - x_2^n}{x_1 - x_2} = \frac{(\frac{1+\sqrt{5}}{2})^n - (\frac{1-\sqrt{5}}{2})^n}{\frac{1+\sqrt{5}}{2} + \frac{-1+\sqrt{5}}{2}} = \frac{(\frac{1+\sqrt{5}}{2})^n - (\frac{1-\sqrt{5}}{2})^n}{\sqrt{5}}$$ $$\blacksquare$$

## $\color{blue}{\text{Pregunta 2}}$

### $\color{blue}{\text{Primera parte: Demuestra que todo árbol con n vértices tiene n − 1 aristas}}$

### $\color{violet}{\text{Demostración:}}\$

Sea G un árbol. Procederemos por inducción sobre el número de vértices, es decir $|V|$.

$\color{violet}{\text{caso base n = 1}}$ \
Se cumple que en efecto el número de aristas es $|V| - 1$ por lo tanto se cumple el caso base\
$\color{violet}{\text{hipótesis de inducción n = k}}$ \
Supongamos que se cumple para alguna k natural de modo que $\text{número de aristas} = |E| = |V| - 1 = k -1$
$\color{violet}{\text{paso de inductivo, por demostrar que se cumple para n = k + 1}}$ 
Supongamos que ahora tenemos un árbol de k + 1 vértices.
Sabemos por lo visto en clase que si cortamos cualquier arista de un árbol, este mismo hace que ahora tengamos dos componentes conexas, es decir, nuestro árbol deja de ser conexo. Veamos a estas componentes conexasa resultantes, dígamosles $G_1$ y $G_2$, entonces tenemos que estas mismas son árboles también (subgráficas de gráficas acíclicas son acíclicas también), siendo así las siguientes $$|E_{G}| = |E_{G_1}| + |E_{G_2}| + 1$$ $$|V_{G}| = |V_{G_1}| + |V_{G_2}|$$, en donde tenemos por hipótesis de inducción que para cualquiera k, se tiene que $|V| - 1 = |E|$ $$|V_{G}| = |V_{G_1}| + |V_{G_2}| = (|E_{G_1}| + 1) + (|E_{G_1}| + 1) = (|E_{G_1}| + |E_{G_2}| + 1) + 1 = |E_G| + 1$$ $$\blacksquare$$

### $\color{blue}{\text{Segunda parte: cualquier bosque con n vértices y k componentes conexas tiene n − k aristas.}}$

### $\color{violet}{\text{Demostración:}}\$

Supongamos que tenemos k componentes conexas con órden $n_1,...,n_k$ respectivamente. Siendo así tenemos $\sum_{i = 1}^k n_i = n$ entonces cada componente es un árbol, en donde cada tenemos $n_i - 1$ aristas. De modo que el número total de aristas es $$\sum_{i=1} ^k (n_i - 1) = \sum_{i=1}^k n_i - \sum_{i=1}^k 1 = n - k$$ por lo tanto tenemos que en un bosque con n vértices y k componentes conexas tenemos n - k aristas $$\blacksquare$$

## $\color{blue}{\text{Pregunta 4}}$
### $\color{blue}{\text{Inciso a) Muestra que |A| = |B|.}}$
### $\color{violet}{\text{Solución}}$
Por hipótesis tenemos que nuestra gráfica es bipartita, en A y B con cada conjunto siendo 3-regular, entonces tenemos $$|E_A| = |V_A|\cdot 3$$ $$|E_B| = |V_B|\cdot 3$$ ahora juntando estas igualdades tenemos: $$|E_A| = |V_A|\cdot 3 = |V_B|\cdot 3 = |E_B|$$ de modo que despejando al 3 obtenemos que $$|V_A| = |V_B|$$ por lo tanto tenemos que $|A| = |B|$ $$\blacksquare$$

### $\color{blue}{\text{Inciso b) Muestra que G tiene un emparejamiento que cubre a A. Como sugerencia, tendrás que usar el teorema de Hall.}}$
### $\color{violet}{\text{Solución}}$
Por hipótesis, A es 3-regular, por lo cuál supongamos $A_1\subseteq A$ y sea k el número de aristar que terminan en $A_1$, ya que todo vértice en $A_1$ tiene grado 3, se tiene que $3\cdot |V_{A_1}| = k$. Sea $N(A_1)$ el conjunto de vecinos de $A_1$ es decir, los vértices que son adyacentes en $A_1$, entonces $|V_{A_1}|$ es a lo más $|N(A_1)|$, por el teorema de Hall hay un emparejamiento que cubre a A, que por corolario, es un emparejamiento perfecto. $$\blacksquare$$
